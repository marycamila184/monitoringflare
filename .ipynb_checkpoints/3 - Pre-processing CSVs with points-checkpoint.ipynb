```python
import pandas as pd
import glob
import os
import pandas as pd
import h5py
import glob
from datetime import datetime, timedelta
import math
import matplotlib.pyplot as plt
import numpy as np
from pyproj import Proj
from collections import OrderedDict
import math
from math import radians
```

```python
joined_files = os.path.join(".\\linkscsvs_points", "*.csv")
joined_list = glob.glob(joined_files)
df = pd.concat(map(pd.read_csv, joined_list), ignore_index=True)
df
```

```python
dfCoordinates = df[df['Lon_GMTCO'] != 999999]
dfCoordinates = dfCoordinates[dfCoordinates['Lat_GMTCO'] != 999999]
```

## (will keep all points for testing) Removing null temperatures and removing temperature below 1400 K based on articles describing a flare activity

```python
dfCoordinates = dfCoordinates[dfCoordinates['Temp_BB'] != 999999]
#dfCoordinates = dfCoordinates[dfCoordinates['Temp_BB'] >= 1400]
```

```python
dfCoordinates['Date_Mscan'] = pd.to_datetime(dfCoordinates['Date_Mscan'], format='%Y/%m/%d %H:%M:%S')
dfCoordinates['Date_Mscan'] = dfCoordinates['Date_Mscan'].dt.date
dfCoordinates.groupby('Date_Mscan').size().reset_index(name='Count').sort_values('Count', ascending=False).tail(15)
```

## Removing duplicated coordinates

```python
dfCoordinates.drop_duplicates(subset=['Lat_GMTCO', 'Lon_GMTCO'], keep= 'last', inplace=True)
dfCoordinates.info()
```

## Checking the period with more detections

```python
dfCoordinates['Date_Mscan'] = pd.to_datetime(dfCoordinates['Date_Mscan'], format='%Y/%m/%d %H:%M:%S')
dfCoordinates.head()
```

```python
dfCoordinates['Date_Mscan'] = dfCoordinates['Date_Mscan'].dt.date
dfCoordinates.groupby('Date_Mscan').size().reset_index(name='Count').sort_values('Count', ascending=False).tail(15)
```

## Getting tiles

```python
def get_tile(lat, lon):
    h = str(math.trunc(abs(((-180 - (lon))/10))))
    v = str(math.trunc(abs(((90 - (lat))/10))))
    
    return "h"+ h.zfill(2)+"v"+v.zfill(2)
```

```python
dfCoordinates['tile'] = dfCoordinates.apply(lambda x: get_tile(x['Lat_GMTCO'], x['Lon_GMTCO']), axis=1)
dfCoordinates.head()
```

```python
dfCoordinates = dfCoordinates[['Lat_GMTCO','Lon_GMTCO', 'Temp_BB','Date_Mscan','tile']]
dfCoordinates.info()
```

```python
dfCoordinates.to_csv('.\\rawpoints_coordinates\\main_coordinates_processed.csv', index=False)
```

## Grouping coordinates by day

```python
df_det = pd.read_csv('.\\rawpoints_coordinates\\main_coordinates_processed.csv')
df_det['Date_Mscan'] = pd.to_datetime(df_det['Date_Mscan'], format='%Y-%m-%d')
df_det
```

```python
init_date = datetime.strptime('2019-08-01', '%Y-%m-%d')
end_date = datetime.strptime('2019-08-31', '%Y-%m-%d')

day_csv = 213

while init_date.date() <= end_date.date():
    df_dates_det_D = df_det[df_det['Date_Mscan'] == init_date]
    df_dates_det_D.to_csv(".\\rawpoints_coordinates\\daily\\A2019" + str(day_csv) + ".csv", index=True)
    init_date = init_date + timedelta(days=1)
    day_csv+=1
```
